```python
import pandas as pd
import tensorflow
from tensorflow import keras
from keras.models import model_from_json
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
import sklearn.metrics

base = pd.read_csv('iris.csv')

entradas = base[base.columns[0:4]].to_numpy()
saidas = base[base.columns[4]]

classesSaida = {'Iris-setosa': 0, 'Iris-versicolor': 1, 'Iris-virginica': 2}
saidasInt = np.empty((len(saidas), 1), dtype=int)
for i in range(len(saidas)):
  saidasInt[i] = classesSaida[saidas[i]]

camadas_escondidas = [1, 5, 7, 10]
neuronios_por_camada = [25, 50, 100, 200, 500]

entradasTrain, entradasTest, saidasTrain, saidasTest = train_test_split(entradas, saidasInt, test_size=0.3)
modelos_config = {}
resultados_por_modelo = {}
for camada in camadas_escondidas:
  for neuronios in neuronios_por_camada:
    modelo = keras.Sequential()
    for i in range(camada):
      modelo.add(keras.layers.Dense(neuronios, activation=tensorflow.nn.relu))
    modelo.add(keras.layers.Dense(3, activation=tensorflow.nn.softmax))
    modelos_config[(camada, neuronios)] = modelo

i = 1
for config in modelos_config:
    print(f"Testando modelo {i}/{len(modelos_config)}")
    i+=1
    modelo = modelos_config[config]
    modelo.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics='accuracy')

    modelo.fit(entradasTrain, saidasTrain, epochs=100, batch_size=32, verbose=1, validation_data=(entradasTest, saidasTest))

    perda_teste, acuracia_teste = modelo.evaluate(entradasTest, saidasTest)
    resultados_por_modelo[config] = acuracia_teste

i = 1
for modelo in resultados_por_modelo:
   print(f"Configurao {i}/{len(resultados_por_modelo)} -> Camadas escondidas: {modelo[0]}, Neurnios: {modelo[1]}, Acurcia: {resultados_por_modelo[modelo]}")
   i+=1
```
